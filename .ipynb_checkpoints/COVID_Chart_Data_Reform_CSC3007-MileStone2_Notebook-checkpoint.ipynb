```python
import os,pandas as pd,numpy as np, json
dfArray=[]
last=[]
dates=[]
folderPath="C:\\Users\\I\\Desktop\\covid_multi_line_chart_CSC3007_milestone2\\jhu_csvs"
directory = os.path.join(folderPath)
count=1
for root,dirs,files in os.walk(directory):
    for file in files:
        if (file.endswith(".csv")):
            df = pd.read_csv(folderPath+"\\"+file)
            try:
                df1=df[["Country_Region","Confirmed","Deaths"]]
            except:
                pass
            
            try:
                df1=df[["Country/Region","Confirmed","Deaths"]]
            except:
                pass
            df1.rename(columns={"Country_Region":"Country"},errors='ignore',inplace=True)
            df1.rename(columns={"Country/Region":"Country"},errors='ignore',inplace=True)
#             countriesNotFound= list(set(allCountries) - set(df1.Country.values))
#             for country in countriesNotFound:
#                 df1.loc[len(df1)] = [country,0,0]
                
            df1=df1.groupby("Country",as_index=False).sum()
            #extra code for testing
#             df1.drop(df1[~df1.Country.isin(["Australia","Antarctica","Cambodia"])].index, inplace=True)
            #resume code
            df1["date"]=str(file.split(sep=".")[0])
            dates.append(file.split(".")[0])
            
            df1.drop(df1[df1.Country.isin(dropCountries)].index, inplace=True)
            dfArray.append(df1)
            
            if (file.split(".")[0]=="07-06-2022"):
                last.append(df1)
            count+=1
```
